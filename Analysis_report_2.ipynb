#Converted 0 and 1 values of senior citizen to yes/no to make it easily to understand

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
telco['SeniorCitizen'] = telco["SeniorCitizen"].apply(conv)        
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a crosstab of counts
count_data = pd.crosstab(telco["SeniorCitizen"], telco["Churn"])

# Step 2: Convert counts to percentages
percentage_data = count_data.div(count_data.sum(axis=1), axis=0) * 100

# Step 3: Plot
ax = percentage_data.plot(kind="bar", stacked=True, figsize=(6, 6), colormap="Set2")

# Step 4: Add percentage labels
for i, total in enumerate(percentage_data.values):
    cum_sum = 0
    for j, val in enumerate(total):
        height = val
        if height > 0:
            ax.text(i, cum_sum + height / 2, f"{height:.1f}%", ha='center', va='center', fontsize=9)
            cum_sum += height

plt.title("Churn by SeniorCitizen (%)")
plt.ylabel("Percentage")
plt.xlabel("Senior Citizen")
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Number of columns for subplot grid (adjust as needed)
n_cols = 3
n_rows = -(-len(columns) // n_cols)  # Ceiling division

# Set figure size
plt.figure(figsize=(18, n_rows * 4))

# Create subplots
for idx, col in enumerate(columns, 1):
    plt.subplot(n_rows, n_cols, idx)
    sns.countplot(data=telco, x=col, hue=telco["Churn"])
    plt.title(f'Count Plot of {col}')
    plt.xticks(rotation=30)
    plt.xlabel(col)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

<h3>#Customers with PhoneService = Yes and MultipleLines = No form the largest groups, but those with MultipleLines = Yes show relatively higher churn rates.

#Services like OnlineSecurity, OnlineBackup, DeviceProtection, and TechSupport show significantly higher churn among those who don’t have these services.

#Customers with Fiber optic InternetService exhibit a higher churn rate compared to DSL or no internet service.

#Features like StreamingTV and StreamingMovies show a noticeable trend of higher churn among users who use these services.</h3>
